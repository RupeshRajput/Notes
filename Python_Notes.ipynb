##### **Use Separator(sep) Parameter**

- Separator Default Value - Space(' ')

```python
print("Apple","Banana","Cherry")
```

```python
print("Apple","Banana","Cherry" ,sep=" ")
print("Apple","Banana","Cherry" ,sep="/")
```

```python
print("16","Mar","2026",sep="-")
```

#### ----------------------------------------------------------------------

##### **Use End Argument**

- End Default Value - Newline(\n)

```python
print("Apple","Banana","Cherry" ,sep=" ",end="$$$")
print("Apple","Banana","Cherry" ,sep="/")
```

```python
print("this is \\n krish\na")
```

## **Python Data Types :**
Python mein data types ye batate hain ki variable kis tarah ki value store kar raha hai. Python ek Dynamically Typed language hai, iska matlab humein variable ka type manually declare karne ki zarurat nahi padti.

#### ***A. Numeric Types (Numbers)***
- **int:** Integer values (e.g., 5, -10).
- **float:** Decimal values (e.g., 3.14, 2.0).
- **complex:** Complex numbers (e.g., 1 + 2j).

#### ***B. Sequence Types (Ordered Collections)***
- **str (String):** Text data (e.g., "Apple" jo aapne code mein likha hai).
- **list:** Mutable (changeable) list (e.g., [1, 2, 3]).
- **tuple:** Immutable (fixed) list (e.g., (1, 2, 3)).

#### ***C. Mapping Type***
- **dict (Dictionary):** Key-value pairs (e.g., {"name": "Apple", "age": 25}).

#### ***D. Set Types***
- **set:** Unique elements ka collection (e.g., {1, 2, 3}).
- **frozenset:** Immutable version of set.

#### ***E. Boolean Types***
- **bool:** Sirf do values - True ya False.
